```python
def my_bin_2_dec(b):
    d = 0
    for i in range(len(b)):
        d += 2**i * int(b[-1 - i])
    return d
```

Problem 2

```python
def my_dec_2_bin(d):
    b = []

    while d > 0:
        b.append(d % 2)
        d //= 2

    b = b[::-1]
    return b
```
