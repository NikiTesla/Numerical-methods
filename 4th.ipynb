Problem 1

```python
def my_nth_root(x, n, tol):
    f = lambda y: y ** n - x
    df = lambda y: n * y ** (n-1)

    x0, i = 1, 0
    R = [x0]
    E =  [abs(f(x0))]

    while E[i] > tol:
        i += 1
        x1 = x0 - f(x0) / df(x0)
        R.append(x1)
        E.append(abs(f(x1)))
        x0 = x1
    return R, E
```

```python
R, E = my_nth_root(30, 3, 1e-3)

print(R, E)
```

```python
def my_newton(f, df, x0, tol):
    R = [x0]
    E = [abs(f(x0))]
    i = 0
    
    while E[i] > tol:
        i += 1
        x1 = x0 - f(x0) / df(x0)
        x0 = x1
        R.append(x1)
        E.append(abs(f(x1)))

    return R, E
```

```python
f = lambda x: x**2 - 2
df = lambda x: 2*x
R, E = my_newton(f, df, 1, 1e-5)

print(R, E)
```

```python
f = lambda x: np.sin(x) - np.cos(x)
df = lambda x: np.cos(x) + np.sin(x)
R, E = my_newton(f, df, 1, 1e-5)
print(R, E)
```
