```python
import numpy as np
```

```python
def f(x):
    return np.exp(x)

def f0(x):
    return np.exp(x)

def my_int_calc(f, f0, a, b, n, option):
    x = []
    I = 0
    dx = (b - a) / (n - 1)
    
    for i in range(n):
        x.append(a + i * dx)

    if option == 'trap':
        for i in range(1, n-1):
            I += f(x[i])

        I *= 2
        I += f(x[0]) + f(x[n - 1])
        I *= dx * 0.5
        
    if option == 'rect':
        for i in range(1, n):
         I += f(x[i]) * dx
         
        
    if option == 'simp':
        for i in range(1, n-1):
            I += f(x[i])
        I *= 4
        for i in range(2, n-2, 2):
            I += 2 * f(x[i])
        I += f(x[0]) + f(x[n - 1])
        I *= dx / 3
        
    J = (f0(b) - f0(a))
    

    return I, J, abs(I-J)
```

```python
print('trap/rect/simp')
name = input()

print('Enter a: ')
a = int(input())

print('Enter b: ')
b = int(input())

print('Enter n: ')
n = int(input())
```

```python
try:
    [I, J, errI] = my_int_calc(f, f0, a, b, n, name)
except ZeroDivisionError:
    print('You divided by zero')
except:
    print('Mistake! Please, try again')
```

```python
print("Numerical: ", I)

print("Analitic: ", J)

print("Error: ", errI)
```
